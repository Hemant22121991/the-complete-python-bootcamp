```python
def myfunc(a,b):
    # returnss 5% of the sum of a and b
    return sum((a,b)) * 0.05
```

```python
myfunc(50,60)
```

```python
def myfunc(*args):
    return sum(args) *0.05
```

```python
myfunc(10,50,60,150,151,51,6,51)
```

```python
def myfunc(*spam):
    return sum(spam) *0.05
```

```python
myfunc(10,50,60,150,151,51,6,51)
```

```python
def myfunc(*args):
    for item in args:
        print(item)
```

```python
def myfunc(**kwargs):
    if 'fruit' in kwargs:
        print(kwargs)
        print('My fruit of choice is {}'.format(kwargs['fruit']))
    else:
        print('I did not find any fruit here')
```

```python
myfunc(fruit = 'apple', veggie = 'lettuce')
```

```python
def myfunc(**jelly):
    if 'fruit' in jelly:
        print(jelly)
        print('My fruit of choice is {}'.format(jelly['fruit']))
    else:
        print('I did not find any fruit here')
```

```python
myfunc(fruit = 'apple', veggie = 'lettuce')
```

```python
def myfunc(*args, **kwargs):
    print(args)
    print(kwargs)
    print('I would like {} {}'. format(args[0], kwargs['food']))
```

```python
myfunc(10,20,30,fruit = 'orange', food = 'eggs', animal = 'dog')
```

```python
def test_func(a):
    mylist = list(a)
    mydict = {}
    fnl_list = []
    for i in mylist:
        # idx = mylist.index(i) + 1
        if (mylist.index(i) + 1)%2 == 0:
            mydict[mylist.index(i) + 1] = i.upper()
        else:
            mydict[mylist.index(i) + 1] = i.lower()
    for key,value in mydict.items():
        fnl_list.append(value)
    print(''.join(fnl_list))
```

```python
test_func('Hemant')
```

```python

def my_func(*args):
    word = args[0]
    print(word)
    result = ''.join(
        ch.lower() if idx % 2 == 0 else ch.upper()
        for idx, ch in enumerate(word)
    )

    return result
```

```python
my_func('Hemant')
```

```python
word = 'Hemant'
result = []

for idx, ch in enumerate(word):
    if idx % 2 == 0:
        result.append(ch.lower())
    else:
        result.append(ch.upper())

print(''.join(result))
```
