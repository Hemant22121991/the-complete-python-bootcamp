```python
# Return true if any number is even inside a list
```

```python
def check_even_list(num_list):
    numbers = []
    for number in num_list:
        if number%2 == 0:
            numbers.append(number)

    return numbers
```

```python
check_even_list(list(range(1,10)))
```
