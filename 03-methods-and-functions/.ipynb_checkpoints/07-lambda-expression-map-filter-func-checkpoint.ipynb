```python
def sqr(num):
    return num ** 2
```

```python
my_nums = [1,2,3,4,5]
```

```python
for item in map(sqr, my_nums):
    print(item)
```

```python
list(map(sqr, my_nums))
```

```python
def splicer(mystring):
    if len(mystring)%2 == 0:
        return 'EVEN'
    else:
        return mystring[0]
```

```python
names = ['Andy', 'Eve', 'Sally']
```

```python
list(map(splicer, names))
```

```python
map(sqr(), my_nums)
```

```python
def check_even(num):
    return num%2 == 0
```

```python
mynums = [1,2,3,4,5,6]
```

```python
list(filter(check_even, mynums))
```

```python
for n in filter(check_even, mynums):
    print(n)
```

```python
# lamda experssion
```

```python
def sqr(num):
    result = num ** 2
    return result
```

```python
sqr(3)
```

```python
def sqr(num): return num ** 2
```

```python
sqr(3)
```

```python
sqr = lambda num: num ** 2
```

```python
sqr(3)
```

```python
# Lamda is anonymous function
mynums = [1,2,3,4,5,6]
```

```python
list(map(lambda num: num ** 2, mynums))
```

```python
list(filter(lambda num: num%2 == 0, mynums))
```

```python
list(map(lambda num: num%2 == 0, mynums))
```
