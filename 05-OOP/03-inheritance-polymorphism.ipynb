```python
# Base class
class Animal():

    def __init__(self):
        print("Animal Created")

    def who_am_i(self):
        print("I am an Animal")

    def eat(self):
        print("I eat chapati Bhaji!")
```

```python
myanimal = Animal()
```

```python
myanimal.who_am_i()
```

```python
myanimal.eat()
```

```python
# Derived Class
class Dog(Animal):
    
    def __init__(self):
        Animal.__init__(self)
        print("Dog Created")
    
    
```

```python
myanimal = Animal()
```

```python
mydog = Dog()
```

```python
mydog.eat()
```

```python
class Dog(Animal):
    
    def __init__(self):
        Animal.__init__(self)
        print("Dog Created")

    def eat(self):
        print("Now I am eating anda-bhurji")
        
    def bark(self):
        print("Woof!")
```

```python
mydog = Dog()
```

```python
mydog.bark()
```

```python
# Polymorphism

class Dog():
    def __init__(self, name):
        self.name = name

    def speak(self):
        return self.name + " says Woof!"
```

```python
class Cat():
    def __init__(self, name):
        self.name = name

    def speak(self):
        return self.name + " says meow!"
```

```python
niko = Dog("Niko")
```

```python
felix = Cat("Felix")
```

```python
print(niko.speak())
```

```python
print(felix.speak())
```

```python
for pet in [niko, felix]:
    print(pet)
    print(pet.speak())
```

```python
def pet_speak(pet):
    print(pet.speak())
```

```python
pet_speak(niko)
```

```python
pet_speak(felix)
```
