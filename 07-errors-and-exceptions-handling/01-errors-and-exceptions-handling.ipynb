```python
def ask_for_int():
    
    while True:
        try:
            result = int(input("Please provide number: "))
        except:
            print("Wront Input. That is not an integer")
            continue
        else:
            print("Thank You!")
            break
        finally:
            print("End of try/except/finally")
```
